# NLP -Analyzing History.txt

## 1. Perform word Tokenization

```python
#uploading the data file to colab environment
from google.colab import files
uploaded = files.upload()
```

```python
#loading the text file (data)
with open('History_AI.txt', 'r') as f:
  history = f.read()

print(history)
```

```python
#loading the nltk modules
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
```

```python
#downloading the neccessary nltk data files for preprocessing
nltk.download('punkt')
nltk.download('stopwords')
```

```python
#tokenizing the words from the loaded text
tokens = word_tokenize(history)
print(tokens)
```

##### Comment
Looking at the output above, it could be noticed that (<b><i>punctuation</i></b>) is added as words and some commonly used words (<b><i>article, preposition</i></b>), thmat do not add much to the meaning of the sentence are also included. This required preprocessing and stopwords removal

```python
#Preprocessing:

#Convert all to lower case. This helps in ensuring uniformity
history = history.lower()

# Remove punctuation
history = history.translate(str.maketrans('', '', string.punctuation))

#Now, performing the tokenization after preprocessing
tokens = word_tokenize(history)
print(tokens)
```

```python
#Now removing the stopwords

stop_words = set(stopwords.words('english'))

filter_tokens = [ word for word in tokens if word not in stop_words]

print(f"Final Tokenize words \n")
print("====================================================================================================================================\n")
print(filter_tokens)
```

```python
#visualizing the tokens using wordcloud
# Joining the tokens back to a single string for word cloud generation
from wordcloud import WordCloud
import matplotlib.pyplot as plt
filtered_text = ' '.join(filter_tokens)

# Generating a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text)

# Visualize the word cloud
plt.figure(figsize=(8, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Frequency of the Tokenize Words')
plt.show()
```

```python
#visualizing the tokens for better presentation
from collections import Counter
import matplotlib.pyplot as plt

# Count the frequency of each token
token_counts = Counter(filter_tokens)

# Get the most common tokens
most_common_tokens = token_counts.most_common(10)

# Separate the tokens and their counts for plotting
tokens, counts = zip(*most_common_tokens)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(tokens, counts, color='blue')
plt.xlabel('Tokens')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Tokenized Words')
plt.xticks(rotation=45)
plt.show()
```

## 2. Perform text classification.

```python
#NATURAL LANGUAGE PROCESSING:
from transformers import pipeline
import torch

#Using the history.txt
classify_history = history

classifier = pipeline('text-classification', model='distilbert-base-uncased-finetuned-sst-2-english', revision='main')
classification = classifier(classify_history)

print("\nText Classification:", classification)
```

## 3. Perform named entity recognition.

```python
#Loading the library
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

#reloading the data
with open('History_AI.txt', 'r') as f:
  ner_history = f.read()

# Load the model and tokenizer directly
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

# Create the NER pipeline
ner_pipeline = pipeline('ner', model=model, tokenizer=tokenizer)

# Perform NER on the preprocessed text
ner_results = ner_pipeline(ner_history)

print(ner_results)
```

```python
#display the entity in data frame for easier recognition
import pandas as pd

# Create a DataFrame from the NER results
ner_df = pd.DataFrame(ner_results)

# Display the DataFrame
print("\nEntity name Recognistion:\n",ner_df)
```

```python
#showing entity name recognition with barchat
import matplotlib.pyplot as plt
import seaborn as sns

# Extract entity types and their labels
entity_types = [entity['entity'] for entity in ner_results]
entity_labels = {
    'B-MISC': 'Beginning of a Miscellaneous entity',
    'I-MISC': 'Inside a Miscellaneous entity',
    'B-PER': 'Beginning of a Person\'s name',
    'I-PER': 'Inside a Person\'s name',
    'B-ORG': 'Beginning of an Organization\'s name'
}

# Create a DataFrame for visualization
entity_df = pd.DataFrame(entity_types, columns=['Entity Type'])

# Plot the frequency of each entity type
plt.figure(figsize=(8, 6))
sns.countplot(data=entity_df, x='Entity Type', palette='viridis')

# Add the title and labels
plt.title('Frequency of Entity Types', fontsize=16)
plt.xlabel('Entity Type', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Add a legend explaining the entity types
handles = [plt.Rectangle((0, 0), 1, 1, color=sns.color_palette('viridis')[i]) for i in range(len(entity_labels))]
plt.legend(handles, [f"{key}: {value}" for key, value in entity_labels.items()], loc='upper right', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

## 4. Perform summarization

```python
summarizer = pipeline('summarization', model="facebook/bart-large-cnn", revision="main")
summary = summarizer(history, max_length=150, min_length=30, do_sample=False)
print("\nSummary:", summary[0]['summary_text'])
```

## 5. Perform Question Answering

```python
qa = pipeline('question-answering', model="distilbert-base-cased-distilled-squad", revision="main")
questions = [
    "What is AI?",
    "Who developed the Turing test?",
    "What is deep learning?"
]
for question in questions:
    answer = qa(question=question, context=history)
    print(f"\nQuestion: {question}")
    print(f"Answer: {answer['answer']}")
```

```python
# 6. Translate the text to Spanish
translator = pipeline('translation_en_to_es', model="Helsinki-NLP/opus-mt-en-es", revision="main")
translation = translator(history)
print("\nTranslation to Spanish:", translation[0]['translation_text'])
```
